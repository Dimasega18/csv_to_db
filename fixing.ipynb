```python
import pandas as pd
import os
```

```python
df = pd.read_csv('./Worlds_Best_50_Hotels.csv',encoding='latin-1')
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df_new = df.iloc[:,[0,1,2,4,5,8,9,10]]
df_new.isna().sum()
```

```python
df_new['Hotel Ammenties'].fillna('Not Available',inplace=True)
df_new['Number'].fillna('Not Available',inplace=True)
df_new['Starting Rate in ($)'] = df_new['Starting Rate in ($)'].astype(float)
df_new.head()
```

```python
df_new.rename(columns={"Rank":"rank","Name":"name","Location":"location","Total Rooms":"total_rooms","Starting Rate in ($)":"starting_rate_dollar","Hotel Ammenties":"hotel_ammenties","Address":"address","Number":"number"},inplace= True)
df_new.head()
```

```python
df_new.loc[df_new['name'].str.contains('[àéñçôã]',regex=True)]
```

```python
df_new.loc[df_new['location'].str.contains('[àéñçôã]',regex=True)]
```

```python
df_new.loc[df_new['hotel_ammenties'].str.contains('[àéñçôã]',regex=True)]
```

```python
df_new.loc[df_new['address'].str.contains('[àéñçôã]',regex=True)]
```

```python
df_new['name'] = df_new['name'].str.replace('à','a').str.replace('é','e').str.replace('ñ','n').str.replace('ç','c').str.replace('ô','o').str.replace('ã','a').str.replace('','-')
df_new['location'] = df_new['location'].str.replace('à','a').str.replace('é','e').str.replace('ñ','n').str.replace('ç','c').str.replace('ô','o').str.replace('ã','a').str.replace('','-')
df_new['hotel_ammenties'] = df_new['hotel_ammenties'].str.replace('à','a').str.replace('é','e').str.replace('ñ','n').str.replace('ç','c').str.replace('ô','o').str.replace('ã','a').str.replace('','-')
df_new['address'] = df_new['address'].str.replace('à','a').str.replace('é','e').str.replace('ñ','n').str.replace('ç','c').str.replace('ô','o').str.replace('ã','a').str.replace('','-')
df_new.head(5)
```
